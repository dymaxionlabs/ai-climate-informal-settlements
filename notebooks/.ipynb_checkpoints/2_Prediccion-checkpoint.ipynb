# Informal settlements

# 2_ Prediction Notebook

At this stage, the already trained model processes new images in search of the patterns that define the object of interest. For each image in the prediction dataset, it returns an image where each pixel represents the probability of having detected the object of interest.

```python
import os
import glob
import subprocess
from tqdm import tqdm
import geopandas as gpd
from unetseg.predict import PredictConfig, predict
from unetseg.evaluate import plot_data_results
from satproc.postprocess.polygonize import polygonize 
from satproc.filter import filter_by_max_prob
```

### **Variables statement**

```python
#General
YEAR="2021" #year of the training image
YEAR_PREDICT = "2021"
ZONE = "sula"
RESO = "low"
VERSION='final'

#Bucket
BUCKET_IMG = f"gs://dym-quilmes-trucks-temp/I2UD/imagenes/{ZONE}/{RESO}/{YEAR_PREDICT}/*.tif" #CHECK
#Paths
BASE_PATH = "./../"
PATH_PREDICT_IMAGES = f'images/{ZONE}/{RESO}/{YEAR_PREDICT}/'
```

```python
if ZONE == "sula":
    #satproc
    SIZE =130
    STEP_SIZE = 40
    #AOI_PREDICT = "ciudad_san_pedro_sula.geojson"
    
   
    #train
    INFO = "classA"
    #unet
    CLASS_WEIGHT = [1]
    STEP_PER_EPOCH = 100
    N_CLASSES = 1
    INPUT_SIZE=160
    N_CHANNELS=3
    MODEL_NAME= f'UNet_{ZONE}_{RESO}_{INPUT_SIZE}x{INPUT_SIZE}_{str(SIZE)}_{str(STEP_SIZE)}_spe{str(STEP_PER_EPOCH)}_fecha{YEAR}_{INFO}_{VERSION}.h5'
    #post-process
    THR=0.2
    min_area = 50000 #CHECK value if the resolution changes 
                                    
    
if ZONE == "tegu":
    #satproc
    SIZE =256
    STEP_SIZE = 50
    
   #train
    INFO = "classABC"
    #unet
    CLASS_WEIGHT = [0.2,0.3,0.5]
    STEP_PER_EPOCH = 95
    N_CLASSES = 3
    MODEL_NAME= f'UNet_{ZONE}_{RESO}_{INPUT_SIZE}x{INPUT_SIZE}_{str(SIZE)}_{str(STEP_SIZE)}_spe{str(STEP_PER_EPOCH)}_fecha{YEAR}_{INFO}_{VERSION}.h5'
    #post-process
    threshold_A = 130
    threshold_B = 130
    threshold_C = 200
    min_area = 10000  #CHECK value if the resolution changes

if RESO =="low":
    
    RESCALE_MODE = "s2_rgb_extra --lower-cut 0 --upper-cut 100"
if RESO =="planet":
    
    RESCALE_MODE="percentiles --lower-cut 0 --upper-cut 100"
    
```

## Prediction dataset generation

### Satproc

This tool is also used to generate the predict dataset which only has images.

We define the predict dataset and the prediction images paths

```python
path_to_files   = f'{BASE_PATH}{PATH_PREDICT_IMAGES}*.tif' #image folder
output_folder   = f'{BASE_PATH}dataset/data_predict/{ZONE}/{RESO}/{YEAR_PREDICT}/final/{str(SIZE)}_{str(STEP_SIZE)}/' #prediction dataset folder
#vector_file_aoi = f'{BASE_PATH}/data/shp/{AOI_PREDICT}' #aoi vector file
print(path_to_files,output_folder)
```

Here, we are interested in predict over the whole image and without overlap, then the size and step_size must be equals.

```python
#!satproc_extract_chips --help
```

```python
!satproc_extract_chips \
                $path_to_files \
                -o  $output_folder \
                --size $SIZE \
                --step-size $SIZE \
                --rescale \
                --rescale-mode $RESCALE_MODE
```

## Prediction

The prediction configuration is defined. We have to pass the prediction dataset path and the model path.

```python
predict_config = PredictConfig(
                            images_path=output_folder, # Prediction dataset folder
                            results_path=os.path.join(BASE_PATH,'dataset/data_results',ZONE,RESO,YEAR_PREDICT,VERSION,str(SIZE)+"_"+str(STEP_SIZE)), #results folder
                            batch_size=16,
                            model_path=os.path.join(BASE_PATH,'data/weights',MODEL_NAME),  #  ruta al modelo (.h5)
                            height=INPUT_SIZE,
                            width=INPUT_SIZE,
                            n_channels = N_CHANNELS,
                            n_classes  = N_CLASSES,
                            class_weights = CLASS_WEIGHT)
```

Running the prediction process.

```python
predict(predict_config)  # Ejecuta la predicción
```

The prediction results can be visualized. We see the image on the left side and the prediction result on the right side.

```python
plot_data_results(num_samples=5, fig_size=(2, 2), predict_config=predict_config, img_ch =2, n_bands=3)
```

# Post processing 

In this stage, we aplied the following post proccessing methods to the prediction results:

* Threshold filtering:In this process we filter the results images by applying a threshold over the probabilities values, saving those which are higher than the threshold value.
* Polygonized
* Minimum area:We select those polygons which area is bigger than the minimun


### **Post-process for Tegucigalpa**

For each settlement class we applied a threshold and after that we assign a pixel value:
* 100 for class B
* 200 for class A
* 255 for class C
With those pixel values a 1-band image is generated and then the results are polygonized.

```python
#Post-process for Tegucigalpa
#thr = THR

if N_CLASSES==3:

    results_path_ =os.path.join(BASE_PATH,'dataset/data_results',ZONE,RESO,YEAR_PREDICT,'final_model_2',str(SIZE)+"_"+str(STEP_SIZE))
    if DEBUG: print (results_path_)

    dest = os.path.join(BASE_PATH,'dataset/data_results',ZONE,RESO,YEAR_PREDICT,'final_model_2','temp_postprocesamiento',str(SIZE)+"_"+str(STEP_SIZE)) # ruta de destino para nuestra predicción
    if DEBUG: print (dest)

    files = next(os.walk(results_path_))[2]
    new_dir = os.path.join(dest, '1band_3classes','thr_'+str(threshold_A)+'_'+str(threshold_B)+'_'+str(threshold_C)+'/')
    os.makedirs(new_dir, exist_ok=True)
    if DEBUG: print (new_dir)
    for file in files:
        if os.path.splitext(file)[-1]=='xml': continue

        result_dst = os.path.join(results_path_,file)
        output_tif_path = os.path.join(dest, '1band_3classes','thr_'+str(threshold_A)+'_'+str(threshold_B)+'_'+str(threshold_C), file)
        exp = "((B<"+str(threshold_B)+")*(C<"+str(threshold_C)+")*(A >"+str(threshold_A)+")*199 ) + ((C<"+str(threshold_C)+")*(B >"+str(threshold_B)+")*99 )+((B<"+str(threshold_B)+")*(C >"+str(threshold_C)+")*254 ) +1"
        cmd_calc = f'gdal_calc.py -A {result_dst} --A_band=1 -B {result_dst} --B_band=2 -C {result_dst} --C_band=3 --outfile {output_tif_path} --calc="{exp}" --NoDataValue=0'
        subprocess.run(cmd_calc, shell=True)
        print("--",output_tif_path)
    print("<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")    
    output_vrt = os.path.join(dest, '1band_3classes','thr_'+str(threshold_A)+'_'+str(threshold_B)+'_'+str(threshold_C),'tiles.vrt')
    output_tif = os.path.join(dest, '1band_3classes','thr_'+str(threshold_A)+'_'+str(threshold_B)+'_'+str(threshold_C),'*.tif')
    cmd_buildvrt = f'gdalbuildvrt {output_vrt} {output_tif}'
    subprocess.run(cmd_buildvrt, shell=True)
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>") 
    output_shp_path = os.path.join('../dataset/data_results',ZONE,RESO,YEAR_PREDICT,'final_model_2', 'poligonize_classABC_thr_'+str(threshold_A)+'_'+str(threshold_B)+'_'+str(threshold_C)+'_'+ZONE+'_'+YEAR+'_'+RESO+'_'+str(SIZE)+'_'+str(STEP_SIZE)+'_'+VERSION+'.gpkg')
    cmd_polygonize = f'gdal_polygonize.py {output_vrt} {output_shp_path}'
    subprocess.run(cmd_polygonize, shell=True)
```

### Threshold

We apply a probability threshold to each municipality result raster

```python
thr=str(int(THR*10))
path_municipios_tif=glob.glob(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/municipios/*.tif')
new_dir=f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/th0{thr}/'
os.makedirs(new_dir, exist_ok=True)
thr1=str(int(THR*255))
for i in path_municipios_tif: 
    base=os.path.basename(i)
    result_input = os.path.join(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/municipios',base)
    output_tif_path =os.path.join(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/th0{thr}',base)
    exp = "((A >"+thr1+")*200)"
    cmd_calc = f'gdal_calc.py -A {result_input} --A_band=1 --outfile {output_tif_path} --calc="{exp}" --NoDataValue=0'
    subprocess.run(cmd_calc, shell=True)
    print(cmd_calc)
    print("--",output_tif_path)
    print("<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<") 
```

The results are polygonized

```python
thr_path=glob.glob(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/th0{thr}/*.tif')
new_dir=os.path.join(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/poligonize/th0{thr}/')
os.makedirs(new_dir, exist_ok=True)

for j in thr_path:
    base3=os.path.basename(j[:-4])
    base2=os.path.basename(j)
    result_thr=os.path.join(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/temp_postprocesamiento/th0{thr}',base2)
    output_shp_path = os.path.join(f'./../dataset/data_results/{ZONE}/{RESO}/{YEAR_PREDICT}/{VERSION}/poligonize/th0{thr}', f"poligonize_classA_thr_0{thr}_{ZONE}_{RESO}_{YEAR_PREDICT}_{VERSION}_{str(SIZE)}_{str(STEP_SIZE)}_{base3}.gpkg")
    cmd_polygonize = f'gdal_polygonize.py {result_thr} {output_shp_path}'
    subprocess.run(cmd_polygonize, shell=True)
    print("---",output_shp_path)
```
